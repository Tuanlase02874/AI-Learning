```python
import numpy as np
import cv2

# Load an color image in grayscale
img = cv2.imread('image_data/user_name.jpg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
left = 100
top = 100
right = 450
bottom = 500
color = (255, 0, 0) 
# Line thickness of 2 px 
thickness = 2
cv2.rectangle(img_rgb,(left,top),(right,bottom),color,thickness)
plt.imshow(img_rgb)
```

```python
left = 200
top = 200
right = 450
bottom = 500
color = (255, 225, 0) 
# Line thickness of 2 px 
thickness = 5
cv2.rectangle(img_rgb,(left,top),(right,bottom),color,thickness)
plt.imshow(img_rgb)
```

```python
user_name = 'user_name'
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 
# org 
org = (left, top) 
# fontScale 
fontScale = 1
# Blue color in BGR 
color = (255, 0, 0) 
# Line thickness of 2 px 
thickness = 2
# Using cv2.putText() method 
image = cv2.putText(img_rgb, user_name, org, font,  
                   fontScale, color, thickness, cv2.LINE_AA) 
plt.imshow(image)
```
