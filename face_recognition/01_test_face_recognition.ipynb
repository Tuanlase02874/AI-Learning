```python
user_name = 'user_name'
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 
# org 
org = (left, top) 
# fontScale 
fontScale = 1

# Blue color in BGR 
color = (255, 0, 0) 
# Line thickness of 2 px 
thickness = 2
# Using cv2.putText() method 
image = cv2.putText(img_rgb, user_name, org, font,  
                   fontScale, color, thickness, cv2.LINE_AA) 
plt.imshow(image)
```

```python
import numpy as np
import cv2
import face_recognition
import matplotlib.pyplot as plt
%matplotlib inline

# Load an color image in grayscale
img = cv2.imread('image_data/khoimy.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
face_locations = face_recognition.face_locations(img_rgb)
for face in face_locations:
    print("face: ",face)
    top, right, bottom, left = face
    color = (255, 255, 0) 
    # Line thickness of 2 px 
    thickness = 5
    cv2.rectangle(img_rgb,(left,top),(right,bottom),color,thickness)
plt.imshow(img_rgb)
```
